```python
from PyPDF2 import PdfFileReader
pdf_path='/content/drive/MyDrive/csml1030/AMALGAMATION_AIRIA.PDF'
pdf = PdfFileReader(str(pdf_path))
```

```python
#pyPDF2 object 
print(pdf)
```

```python
#metadata - creation date and modified date 
pdf.documentInfo
```

```python
#extract alll pagesas a list 
extracted_page = []
for page in pdf.pages:
  extracted_page.append(page.extractText())
```

#  Text extraction using pdfplumber

```python
import re
import requests
import pdfplumber
import os 
import pandas as pd
from collections import namedtuple
```

```python
pdf_path='/content/drive/MyDrive/week1/pdf1.pdf'
```

```python
all_text = '' # new line
with pdfplumber.open(pdf_path) as pdf:
#returns an instance of odfplumber class 
#instance has pages property which is a list of pdfplumber,Page 
  i=0
  for pdf_page in pdf.pages:
    single_page_text = pdf_page.extract_text()
    #print("Page no:", i, " ",  single_page_text)
    i = i + 1
    # separate each page's text with newline
    try:
      all_text = all_text + '\n' + single_page_text
     
    except:
      print("Error")


#reads till the end, but  gives an error at the end , caught with except 
print(all_text)
```
